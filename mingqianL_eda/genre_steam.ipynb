```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
import plotly.io as pio
import altair as alt
```

```python
df = pd.read_csv('../steam.csv')
df.head()
```

```python
mask = ~df['genres'].str.contains(';')

# Apply the mask to the DataFrame to keep only the rows where 'genres' does not contain ';'
df = df[mask]
```

```python
df.groupby(['genres'])['name'].count().sort_values(ascending=False).reset_index().iloc[:50,]
```

```python
grouped_genres = df.groupby('genres').size().reset_index(name='counts')
genres_to_replace = grouped_genres[grouped_genres['counts'] < 10]['genres'].tolist()

# Add a new column to the original DataFrame with the updated genres
df['new_genre'] = df['genres'].apply(lambda x: 'Others' if x in genres_to_replace else x)

# If you want to replace the original 'genre' column with the 'new_genre' column, use:
df['genres'] = df['new_genre']
df.drop('new_genre', axis=1, inplace=True)

df.groupby(['genres'])['name'].count().sort_values(ascending=False).reset_index()
```

```python
genre_counts = df.groupby('genres').size().reset_index(name='counts')

'''alt.Chart(genre_counts).mark_arc().encode(
    theta='counts',
    color="genres"
)'''

base = alt.Chart(genre_counts.iloc[:10,]).encode(
    theta=alt.Theta("counts:Q", stack=True, sort=alt.EncodingSortField(field='counts:Q',order='descending')),
    radius=alt.Radius("counts", scale=alt.Scale(zero=True, rangeMin=20)),
    color=alt.Color('genres:N', legend=alt.Legend(orient='top')),
    tooltip=["genres:N", "counts"]
).properties(
    height=600,
    width = 500)

base.mark_arc()
```
